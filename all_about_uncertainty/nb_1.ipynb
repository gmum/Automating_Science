```python
import numpy as np
import matplotlib.pyplot as plt
```

A. Uncertainty classification

variability within a given model

```python
x = np.linspace(0, 5, 10)
y = 0.15*x + np.random.random(10)
plt.scatter(x, y)
fit = np.polyfit(x, y, 1)
line_x = np.linspace(-1, 6, 2)
plt.plot(line_x, np.poly1d(fit)(line_x))
plt.show()
```

lack of knowledge or information

```python
plt.figure(figsize=(12, 3))
line_x = np.linspace(0, 5, 50)
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.scatter(x, y)
    fit = np.polyfit(np.concatenate([x, [3]]), np.concatenate([y, [i]]), 10)
    plt.plot(line_x, np.poly1d(fit)(line_x))
plt.show()
```
